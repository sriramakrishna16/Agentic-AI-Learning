#### Pydantic basics : Creating and using Models.


```python
from pydantic import BaseModel
```

```python
from dataclasses import dataclass

@dataclass
class Person():
    name:str
    age:int
    city:str

person = Person(name="krishna",age=25,city=35) #runs 
print(person)    
```

from above , we should able to change datatype but by using BaseModel imported from pydantic , validates the data and throws error if we use wrong datatype or value

```python
class Person1(BaseModel):
    name:str
    age:int
    city:str
person1 = Person1(name="krishna",age=25,city="rjy")
# person1 = Person1(name="krishna",age=25,city=23)  
# it throws error
print(person1) 
```

Using optional in pydantic:
    Allow None as a valid value.
    Validate incoming data.
    Automatic typecasting happens.(according to datatype)

```python
from typing import Optional

class employ(BaseModel):
    name:str
    age:int
    city:str
    salary:Optional[float] = None
    is_active:Optional[bool] = True

a = employ(name="krishna",age=25,city="rjy")
print(a) 
```

```python
from typing import List

class Classroom(BaseModel):
    room_number: str
    students: List[str]
    capacity: int
```

```python
classroom = Classroom (
    room_number="A1",
    # students=["arjun","arya","ram"],
    students=("arjun","arya","ram"),
    capacity=30
)
print(classroom)
```

```python
try:
    invalid_val = Classroom(room_number="1B",students=["ram",234],capacity=30)
except ValueError as e:
    print(e)    
```

in the nested model , i have used datatype for zip_code as int and i gave input as numbers but used inverted commas. And the result is automatically type casted into integer value.

```python
class Address(BaseModel):
    street: str
    city: str
    zip_code: int

class Customer(BaseModel):
    customer_id: int
    name: str
    address: Address #nested model

customer = Customer(
    customer_id = 1,
    name = "ram",
    address={"street":"main street", "city": "Rjy","zip_code":"534340"}
)    
print(customer)
```

#### Pydantic field: Customization and constructions

The Field function in pydantics enhances model fields beyond basic type hints by allowingg you to specify validation rules, default values , aliases and more.

```python
from pydantic import Field,BaseModel

class Item(BaseModel):
    name : str=Field(min_length=2,max_length=50)
    price: float=Field(gt=0,le=1000)
    quantity: int=Field(ge=0)

# item = Item(name="Book", price=-1, quantity=10) #Throws error
item = Item(name="Book", price=24.50, quantity=10) 

print(item)
```
