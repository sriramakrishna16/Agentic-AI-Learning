Chroma is a AI native open-source database focused on developer productivity and happiness. Chroma is licensed under Apache 2.0

to use chroma we have to install langchain_chroma library

```python
## building a sample vectordb
from langchain_chroma import Chroma
from langchain_community.document_loaders import TextLoader
from langchain_ollama import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
loader = TextLoader("speech.txt")
data = loader.load()
data
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 500, chunk_overlap = 0)
splits = text_splitter.split_documents(data)
splits
```

```python
embeddings = OllamaEmbeddings(model="nomic-embed-text")
vectordb = Chroma.from_documents(documents=splits,embedding = embeddings)
vectordb
```

```python
query = "What is Artificial Intelligence?"
doc = vectordb.similarity_search(query)
doc[0].page_content
```

```python
## saving to the disk
vectordb = Chroma.from_documents(documents = splits, embedding = embeddings, persist_directory="./chroma_db")
```

```python
## calling the saved db
db = Chroma(persist_directory = "./chroma_db", embedding_function = embeddings)
docs = db.similarity_search("what is artificial intelligence")
docs[0].page_content
```

```python
### retreiving
retriver = vectordb.as_retriever()
retriver.invoke(query)[0].page_content
```
