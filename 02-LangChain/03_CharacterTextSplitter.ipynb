```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader('speech.txt')
docs = loader.load()
docs
```

now importing charcter text splitter

```python
from langchain_text_splitters import CharacterTextSplitter

text_splitter = CharacterTextSplitter(separator="\n\n",chunk_size = 70 , chunk_overlap= 15)
text_splitter.split_documents(docs)
```

```python
with open("speech.txt") as f:
    speech = f.read()

text_splitter1= CharacterTextSplitter(chunk_size = 70 , chunk_overlap=20)
text = text_splitter1.create_documents([speech])
# print(text[0])
# print(text[1])
print(text)
```
