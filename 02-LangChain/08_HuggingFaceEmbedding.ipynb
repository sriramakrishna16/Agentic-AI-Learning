HuggingFace sentence-transformers is a python framework , one of the embedding models used in the hugging face embedding class.

simply it is a text, sentence and image embedding model.

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name = "all-MiniLM-L6-v2")
```

```python
test = "This is a test document"
query_result =  embeddings.embed_query(test)
query_result
```

```python
len(query_result)
```

```python
doc_results = embeddings.embed_query(text = "this is not a test document")
doc_results
```
